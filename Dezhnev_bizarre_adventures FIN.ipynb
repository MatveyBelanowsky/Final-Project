```python
!pip install folium

import pandas as pd
import matplotlib.pyplot as plt
import folium

df = pd.read_csv('dezhnev_dataset.csv')

#print(df.head()) #Проверяем, правильно ли загрузился файл
#Пропускаем этап с проверкой данных и их очисткой, составлять датасеты самому удобно :)

#Подсчёт остановок по годам:
stops_by_year = df['year_period'].value_counts().sort_index()

print("Количество остановок по годам:")

for year, stops_count in stops_by_year.items():
    print(f"{year} - {stops_count}")

#Сортировка остановок по времени
stops_by_duration = df.sort_values(
    by='duration_months',
    ascending=False
)

print("Остановки по времени пребывания:")

for _, row in stops_by_duration.iterrows():
    print(f"{row['location_name']} - {row['duration_months']} месяцев")


#Создание карты
m = folium.Map(
    location=[63.30075875758426, 169.8272843623213],
    zoom_start=5,
    tiles='CartoDB positron',
    control_scale=True,
)

#Добавление маркеров(в силу того, что 3 из 13 маркеров находятся в западном полушарии они переехали на другом конце карты :/ )
for index, row in df.iterrows():
    popup_html = f"""
    <div style="font-family: Arial; font-size: 13px; width: 220px">
        <b>{row['location_name']}</b><br>
        Год: {row['year_period']}<br>
        Количество месяцев: {row['duration_months']}<br>
    </div>
    """
    folium.Marker(
        icon=folium.Icon(color='darkblue', icon='flag'),
        popup=folium.Popup(popup_html, max_width=250),
        location=[row['latitude'], row['longitude']],
        tooltip=row['location_name'],

    ).add_to(m)

print(f"Добавлено {len(df)} маркеров")
m.save('dezhnev.html')
print("Карта сохранена в dezhnev.html")
```
